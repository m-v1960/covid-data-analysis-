# 5. Data Cleaning 



```python
#5,a.Remove all duplicates observations
covid.drop_duplicates()
```

```python
#5,b.Find missing values in all columns
covid.isnull()
```

```python
#5,c. Remove all observations where continent column value is  missing 
covid.dropna(subset=["continent"])
```

```python
#5, d. Fill all missing values with 0 
covid.fillna(0)
```

# 8. Feature Engineering :
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of
# 'total_deaths' column to 'total_cases'
